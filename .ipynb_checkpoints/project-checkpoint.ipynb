# Tratamento inicial do dataset reduzido

```python
cols = ['usableAreas','parkingSpaces','suites','bathrooms','totalAreas','bedrooms']

for var in cols:
    s_extracted_digits = df[var].str.extract('\[[\']{0,1}(\d*)[\']{0,1}\]').squeeze()
    df[var] = s_extracted_digits.apply(lambda x: float(x) if x.isdigit() else np.nan)
df
```

```python
df.to_csv('reduced_dataset.csv',index=False)
```

```python
#transforming column 'amenities' to list instead a string representation of a list
df['amenities'] = df['amenities'].str.strip('][').str.split(', ').apply(lambda x: [i[1:-1]for i in x])
```

```python
#Getting list of unique amenities
unique_amenities = []
for row in df['amenities']:
    for elem in row:
        if elem not in unique_amenities:
            unique_amenities.append(elem)

unique_amenities  
```
